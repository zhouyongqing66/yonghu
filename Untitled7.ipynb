```python
# 从相关系数热力图中可以看出，X32（用户总套餐价值）和X33（用户主资费套餐）两个特征相关性极强，
因此删除其中一个特征，只保留一个
```

```python
del df['X33']
```

```python
#独热编码
前面为了计算特征之间的相关系数而将X1和X5文本型类别转换为数值型类别
，而单纯的将其数值化会让特征赋予了数字大小含义，
而实际上数字大小并没有意义。所以将这两个特征再进行one-hot编码（独热编码）
```

```python
# 生成哑变量矩阵（独热编码）
temp1 = pd.get_dummies(df['X1'])
temp1
```

```python
temp2 = pd.get_dummies(df['X5'])
temp2
```

```python
# 用哑变量矩阵替换原来的“职业”列
df[temp1.columns] = temp1
df[temp2.columns] = temp2
```

```python
df.drop(columns=['X1', 'X5'], inplace=True)
df.info()
```

```python
# X15、X16、X17等几个特征之间的数值大小相差较大，容易影响目标结果，
#使得一些算法无法学习到其他特征，
#因此必须将数值型数据无量纲化，此处采用标准化的方式处理。
```

```python
re_df = df
```

```python
re_df.info()
```

```python
from sklearn.preprocessing import StandardScaler
#标准化转换器
transfer = StandardScaler()
a = transfer.fit_transform(re_df[['X15','X16','X17','X32','X38','X44','X45','X46']])
re_df = pd.concat([re_df,pd.DataFrame(a,columns=['X15_s','X16_s','X17_s','X32_s','X38_s','X44_s','X45_s','X46_s'])],axis=1)

#删除原有列
re_df.drop(['X15','X16','X17','X32','X38','X44','X45','X46'],axis=1,inplace=True)
```

```python
re_df.drop(columns=['user_id', 'product_no'], inplace=True)
```

```python
re_df.info()
```

```python
from sklearn.model_selection import train_test_split
#划分数据集
x_train,x_test,y_train,y_test = train_test_split(re_df.drop(['label'],axis=1),re_df['label'],test_size=0.2,random_state=6)

```

```python
x_train.info()
```

```python
x_test.info()
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score,classification_report

#逻辑回归模型训练
estimator_log = LogisticRegression()
estimator_log.fit(x_train,y_train)
#模型评估
score_log = estimator_log.score(x_test,y_test)
y_pred = estimator_log.predict(x_test)
report_log = classification_report(y_test,y_pred,labels=[0,1],target_names=['不是5G用户','是5G用户'])
auc_log = roc_auc_score(y_test,y_pred)
print('逻辑回归的准确率为%.4f' % score_log,
     '\n精确率、召回率及F1-score为：\n',report_log,
     '\nAUC指标为%.4f' % auc_log)
```

```python
from sklearn.tree import DecisionTreeClassifier

#决策树模型
estimator_tree = DecisionTreeClassifier()
estimator_tree.fit(x_train,y_train)
#模型评估
score_tree = estimator_tree.score(x_test,y_test)
y_pred = estimator_tree.predict(x_test)
report_tree = classification_report(y_test,y_pred,labels=[0,1],target_names=['不是5G用户','是5G用户'])
auc_tree = roc_auc_score(y_test,y_pred)
print('决策树模型的准确率为%.4f' % score_tree,
     '\n精确率、召回率及F1-score为：\n',report_tree,
     '\nAUC指标为%.4f' % auc_tree)
```
